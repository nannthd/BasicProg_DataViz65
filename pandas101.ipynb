## ลบ missing

Screenshot (383).png

1. axis 
 - แทนที่จะลบแถวที่มี NaN อยู่ในคอลัมน์ใดคอลัมน์หนึ่ง ก็อาจเปลี่ยนเป็นลบคอลัมน์ที่มี NaN อยู่ที่แถวใดแถวหนึ่ง ซึ่งทำได้โดยเติมคีย์เวิร์ด axis เข้าไปเป็น axis=1 ความหมายก็คือพิจารณาเป็นคอลัมน์ๆไป
 - 0 คือ index ลบ Rows ที่มีค่า NULL
 - 1 คือ columns ลบ Columns ที่มีค่า NULL
2. tresh
 - กำหนดให้มีค่าที่ไม่ใช่ค่า NA จำนวนมาก
 - เช่น tresh = 2 คือ เก็บเฉพาะแถวที่มีค่าที่ไม่ใช่ค่า NA อย่างน้อย 2 ค่า
3. how 
 - all ลบค่า NULL ทั้งหมดไม่ว่าจะเป็นแถวหรือคอลัมน์

missing -> null, na (not autorized), NaN (not a number)

## Dictionary-RowOriented

```python
sales = [{'account':'Jones LLC','Jan':150,'Feb':200,'Mar':140},
        {'account':'Alpha Co','Jan':200,'Feb':210,'Mar':215},
        {'account':'Blue Inc','Jan':50,'Feb':90,'Mar':95}]
```

```python
df = pd.DataFrame(sales)
df
```

## Dictionary-ColumnOriented

```python
sales = {'account':['Jones LLC', 'Alpha Co', 'Blue Inc'],
         'Jan': [150,200,50],
         'Feb': [200,210,90],
         'Mar': [140,215,95]}
```

```python
df = pd.DataFrame.from_dict(sales)
df
```

# List-RowOriented

```python
sales = [['Jones LLC',	150,	200,	140],
         ['Alpha Co',	200,	210,	215],
         ['Blue Inc',	50,	90,	95]]
columnname = ['account','Jan','Feb','Mar']
```

```python
df = pd.DataFrame.from_records(sales,columns=columnname)
df
```
